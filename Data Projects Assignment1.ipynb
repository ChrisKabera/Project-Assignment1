```python
import pandas as pd
import seaborn as sns

# Load the iris dataset into a Pandas DataFrame
iris_df = sns.load_dataset('iris')

```

```python
# a. Display the first 5 rows of the DataFrame.

df=pd.DataFrame(iris_df.head())
print(df)
```

```python


print(iris_df.info())
```

```python
 # Display summary statistics for numerical columns.
print(iris_df.describe())
```

```python
# Check for duplicates
print("Number of duplicate rows:", iris_df.duplicated().sum())

# Remove duplicates
iris_df = iris_df.drop_duplicates()

# Confirm duplicates are removed
print("Number of duplicate rows after removal:", iris_df.duplicated().sum())

```

```python
#Danipulation

#Rename the columns of the DataFrame:
iris_df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
print(iris_df.columns)  
```

```python
# Create a new column called 'sepal_area':
iris_df['sepal_area'] = iris_df['sepal_length'] * iris_df['sepal_width']

print(iris_df)
```

```python
# Create a new DataFrame called iris_subset containing only rows where 'petal_length' is greater than 3.5:
iris_subset = iris_df[iris_df['petal_length'] > 3.5].copy()

print(iris_subset)
```

```python
#Reset the index of iris_subset:
iris_subset.reset_index(drop=True, inplace=True)

print(iris_subset.reset_index)
```

```python
#Data Analysis
#Calculate the mean 'petal_width' for each species:
mean_petal_width = iris_df.groupby('species')['petal_width'].mean()
print(mean_petal_width)

```

```python
#Find the maximum 'sepal_length' and the corresponding 'species:
max_sepal_length = iris_df['sepal_length'].max()
max_sepal_species = iris_df.loc[iris_df['sepal_length'] == max_sepal_length, 'species'].iloc[0]
print("Maximum sepal length:", max_sepal_length)
print("Species with maximum sepal length:", max_sepal_species)
```

```python
#Count the number of occurrences of each 'species' in the DataFrame

species_counts = iris_df['species'].value_counts()
print(species_counts)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style of the plot
sns.set_style("whitegrid")

# Plot the histogram
plt.figure(figsize=(6, 3))
sns.histplot(data=iris_df, x='petal_length', hue='species', kde=True)
plt.title('Distribution of Petal Length for Each Species')
plt.xlabel('Petal Length')
plt.ylabel('Frequency')
plt.legend(title='Species')
plt.show()
```

```python
#Data Visualization:

import matplotlib.pyplot as plt
import seaborn as sns

# Set the style of the plot
sns.set_style("whitegrid")

# Plot the scatter plot
plt.figure(figsize=(7, 3))
sns.scatterplot(data=iris_df, x='sepal_length', y='sepal_width', hue='species')
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend(title='Species')
plt.show()
```

```python
# Plot a boxplot for each numerical column in the DataFrame

import matplotlib.pyplot as plt
import seaborn as sns

# Set the style of the plot
sns.set_style("whitegrid")

# Plot boxplots for each numerical column
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris_df)
plt.title('Boxplot for Each Numerical Column')
plt.xlabel('Features')
plt.ylabel('Value')
plt.show()
```

```python
 #Create a pairplot for the DataFrame, with points colored by 'species'.
    
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style of the plot
sns.set_style("whitegrid")

# Create pairplot
plt.figure(figsize=(1, 1))
sns.pairplot(data=iris_df, hue='species')
plt.suptitle('Pairplot of Iris Dataset (Colored by Species)', y=1.02)
plt.show()
```
